```python
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset
df = pd.read_csv("cleaned_sales_data.csv")

# Create groups
large_sales = df[df['DEALSIZE'] == 'Large']['SALES']
small_sales = df[df['DEALSIZE'] == 'Small']['SALES']
```

```python
t_stat, p_value = ttest_ind(large_sales, small_sales)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
print(df['DEALSIZE'].unique())
```

```python
alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis (H₀)")
    print("There is a statistically significant difference in average sales between Large and Small deals.")
else:
    print("Fail to Reject the Null Hypothesis (H₀)")
    print("There is no statistically significant difference in average sales between Large and Small deals.")
```

```python
print("Average Sales (Large Deals):", large_sales.mean())
print("Average Sales (Small Deals):", small_sales.mean())
```
